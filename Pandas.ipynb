```python
# Merge,Join&Concatenate DataFrames
```

```python
df1 = pd.DataFrame({
    'id': [1,2,3,4],
    'name':['Keerti','Amit','Educosys','YouTube']
})
df2=pd.DataFrame({
    'id':[3,4,5,6],
    'course':['GenAI','cpp','ds','cs']
})
inner_result=pd.merge(df1,df2,how='inner',on='id')
inner_result
```

```python
outer_result=pd.merge(df1,df2,how='outer',on='id')
outer_result
```

```python
left_result=pd.merge(df1,df2,how='left',on='id')
left_result
```

```python
right_result=pd.merge(df1,df2,how='right',on='id')
right_result
```

```python
 df3=pd.DataFrame({
    'id1':[1,2,3,4],
     'name':['Keerti','Amit','Educosys','YouTube']
 })
df4=pd.DataFrame({
    'id2':[3,4,5,6],
    'course':['GenAI','CPP','HLD','LLD']
})
inner_result=pd.merge(df3,df4,how='inner',left_on='id1',right_on='id2')
inner_result
```

```python
df1 = pd.DataFrame({
    'id': [1,2,3,4],
    'name':['Keerti','Amit','Educosys','YouTube']
})
df2=pd.DataFrame({
    'id':[3,4,5,6],
    'course':['GenAI','cpp','ds','cs']
})
result=pd.concat([df1,df2],axis=0,ignore_index=True)
result
```

```python
result=pd.concat([df1,df2],axis=1)
result
```
